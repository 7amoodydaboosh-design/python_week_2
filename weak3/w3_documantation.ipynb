````python
# Activity 1: Identify the Components

### 1.1. What are the inputs?
* Three quiz marks entered by the user (`quiz1`, `quiz2`, `quiz3`).
* A user decision/input to determine whether to continue entering marks for another student (e.g., 'yes' or 'no').

### 1.2. What is the process?
* Prompting the user to enter data and converting the string inputs into floats.
* Calculating the average mark using the formula: Average = (Quiz 1 + Quiz 2 + Quiz 3) / 3.
* Evaluating whether the average is greater than or equal to 50 to decide if the student passes or fails.
* Employing a conditional loop (`while True`) to repeat the process until the user decides to stop.

### 1.3. What is the output?
* The calculated average mark displayed to two decimal places.
* A status message indicating a "Pass" or "Fail".

---

# 2: Design the Algorithm

### 2.1. Pseudocode
```text
START
    WHILE true DO
        PROMPT "Enter marks for Quiz 1: "
        READ quiz1
        PROMPT "Enter marks for Quiz 2: "
        READ quiz2
        PROMPT "Enter marks for Quiz 3: "
        READ quiz3
        
        IF inputs are not numeric THEN
            DISPLAY "Invalid input!"
            CONTINUE to next loop iteration
        ENDIF
        
        COMPUTE average = (quiz1 + quiz2 + quiz3) / 3
        DISPLAY "The average mark is: ", average
        
        IF average >= 50 THEN
            DISPLAY "Result: Pass"
        ELSE
            DISPLAY "Result: Fail"
        ENDIF
        
        PROMPT "Do you want to enter marks for another student? (yes/no): "
        READ another
        
        IF another is not equal to "yes" THEN
            BREAK loop
        ENDIF
    ENDWHILE
END
````
