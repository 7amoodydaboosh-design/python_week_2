# Activity 1: Identify the Components

### 1.1. Inputs
* Three quiz marks (`quiz1`, `quiz2`, `quiz3`).
* User choice to continue (`yes/no`).

### 1.2. Process
* Get the three marks from the user.
* Calculate average: `(quiz1 + quiz2 + quiz3) / 3`.
* Check if average `>= 50` for pass or fail.
* Loop to allow another student.

### 1.3. Output
* The average mark.
* Pass or Fail message.

---

# Activity 2: Design the Algorithm

### 2.1. Pseudocode
```text
START
    WHILE True
        INPUT quiz1, quiz2, quiz3
        average = (quiz1 + quiz2 + quiz3) / 3
        PRINT average
        
        IF average >= 50 THEN
            PRINT "Pass"
        ELSE
            PRINT "Fail"
        ENDIF
        
        INPUT choice
        IF choice != "yes" THEN
            BREAK
        ENDIF
    ENDWHILE
END

### 2.2. Identify the Algorithm Structure

* **2.2.1. Sequence:**
  The step-by-step sequential execution of prompting for input data, executing the mathematical calculation for the average mark, and outputting the numeric result to the user.

* **2.2.2. Selection:**
  The `if-else` structural statements used to handle error trapping (try-except) for invalid inputs and to check if the student's average qualifies for a **Pass** ($\ge 50$) or a **Fail** ($< 50$).

* **2.2.3. Iteration:**
  The `while True` loop container that allows the entire grade calculation logic to continuously execute for multiple students until the user explicitly inputs anything other than `yes` at the continuation prompt.
